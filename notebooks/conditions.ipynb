# Conditional Statement (Percabangan) Python

Percabangan memungkinkan program memilih aksi berbeda berdasarkan syarat tertentu. Hal ini penting saat logika program harus merespon keadaan yang berubah. Tiga kata kunci utama yang dipakai dalam Python adalah:
- `if`: Mengevaluasi suatu kondisi, dan menjalankan blok kode jika bernilai `True`.
- `elif`: (else if) Mengecek kondisi tambahan ketika kondisi sebelumnya `False`.
- `else`: Menangani semua kondisi yang tidak terpenuhi oleh `if` maupun `elif`.


## Struktur Dasar `if`

Kata kunci `if` mengevaluasi sebuah kondisi bernilai boolean. Jika hasilnya `True`, blok kode di dalamnya dieksekusi. Gunakan struktur ini ketika Anda ingin menjalankan aksi hanya pada kondisi tertentu.

**Contoh kode:** lihat sel kode `Struktur dasar if` tepat setelah penjelasan ini.

### Menambah `else`

Gunakan `else` untuk menyiapkan aksi alternatif ketika kondisi awal `False`.

**Contoh kode:** sel `Menggunakan if-else` memperlihatkan bagaimana menawarkan program membership.

### Menambah `elif`

`elif` (else-if) memungkinkan pemeriksaan beberapa kondisi secara berurutan.

**Contoh kode:** sel `Rantai if-elif-else` menunjukkan cara menentukan dokumen identitas sesuai usia.

Python berhenti pada cabang pertama yang bernilai `True`, jadi urutkan kondisi dari yang paling spesifik ke yang paling umum. Pastikan indentasi konsisten karena Python memakai indentasi untuk menentukan blok kode.


````python
suatu_angka = int(input("Masukkan angka: "))

if suatu_angka % 7 == 0:
    print(f"{suatu_angka} adalah kelipatan dari 7")
````

````python
# Struktur dasar if
stok_garam = 2

if stok_garam < 3:
    print("Pesan ulang garam ke supplier")
````

````python
# Menggunakan if-else
pelanggan_member = False

if pelanggan_member:
    print("Berikan diskon 10%")
else:
    print("Tawarkan program membership")
````

````python
# Rantai if-elif-else
umur = 18

if umur >= 17:
    print("Boleh membuat KTP")
elif umur >= 13:
    print("Boleh membuat Kartu Pelajar SMP")
else:
    print("Belum memenuhi syarat dokumen")
````

## Operator yang Umum Dipakai

| Operator | Nama | Contoh | Keterangan Singkat |
| --- | --- | --- | --- |
| `==` | Sama dengan | `umur == 18` | `True` jika nilai sama persis |
| `!=` | Tidak sama dengan | `kota != 'Makassar'` | `True` jika nilai berbeda |
| `>` | Lebih besar | `nilai > 75` | `True` jika nilai lebih besar dari 75 |
| `>=` | Lebih besar atau sama dengan | `total >= 50000` | Memeriksa batas minimal |
| `<` | Lebih kecil | `stok < 10` | Mengetahui stok menipis |
| `<=` | Lebih kecil atau sama dengan | `jam <= 22` | Membatasi jam layanan |
| `and` | Dan | `pelanggan == 'baru' and total > 100000` | Semua kondisi harus `True` |
| `or` | Atau | `wilayah == 'kota' or ongkir <= 20000` | Salah satu kondisi `True` |
| `not` | Negasi | `not is_member` | Membalik nilai boolean |


## Nested `if` (Percabangan Bersarang)

Nested `if` muncul ketika di dalam satu cabang `if` terdapat percabangan lagi. Teknik ini berguna saat keputusan kedua hanya boleh dievaluasi jika syarat pertama terpenuhi.

**Contoh kode:** sel `Nested if` di bawah menilai jenis pelanggan dan riwayat tagihan.

Tips:
- Gunakan nested `if` hanya ketika logika lanjutan benar-benar bergantung pada cabang sebelumnya.
- Pertimbangkan penulisan ulang memakai operator logika (`and`, `or`) bila struktur bersarang terlalu dalam agar kode tetap mudah dibaca.


````python
# Nested if
pelanggan = {
    "tipe": "korporat",
    "tagihan_lancar": True,
}

if pelanggan["tipe"] == "korporat":
    if pelanggan["tagihan_lancar"]:
        print("Berikan termin pembayaran 30 hari")
    else:
        print("Minta pelunasan sebelum kirim barang")
else:
    print("Gunakan aturan pelanggan umum")
````

## `match` Case (Switch-Case versi Python)

Sejak Python 3.10 tersedia pattern matching dengan kata kunci `match` yang berperilaku mirip switch-case di bahasa lain. Cocok dipakai saat keputusan bergantung pada nilai satu variabel atau pola tertentu.

**Contoh kode:** sel `match-case status pesanan` memperlihatkan cara menampilkan pesan berbeda berdasarkan status logistik.

Catatan penting:
- Gunakan `case _:` sebagai default ketika tidak ada pola lain yang cocok.
- Bisa mencocokkan lebih dari sekadar string, misalnya tuple atau struktur dictionary sederhana.
- `match` membantu menggantikan rantai `if-elif` panjang yang hanya memeriksa kesetaraan nilai tertentu.


````python
# match-case status pesanan
kode_status = "proses"

match kode_status:
    case "baru":
        print("Pesanan diterima, hubungi kurir")
    case "proses":
        print("Sedang diproses di gudang")
    case "dikirim":
        print("Pesanan dalam perjalanan")
    case _:
        print("Status tidak dikenal, cek sistem")
````

## Contoh Studi Kasus: Promo Pengantaran Makanan

Sebuah layanan pesan-antar ingin memberikan promo berdasarkan total belanja, metode pembayaran, dan jam pemesanan. Logika berikut menentukan pesan yang tampil di aplikasi pelanggan.

**Contoh kode:** sel `Promo pengantaran makanan` menghitung promo yang berlaku.

Penjelasan:
- Kondisi pertama memastikan promo utama hanya berlaku bila total belanja tinggi dan pemesanan sebelum jam 22.
- Blok `if` di dalamnya memeriksa metode pembayaran digital tertentu.
- Cabang `elif` menangani pelanggan dengan belanja menengah.
- Cabang `else` menjadi fallback ketika tidak ada promo yang bisa diberikan.


````python
# Promo pengantaran makanan
total_belanja = 175_000
metode_bayar = "ovo"
estimasi_jam = 21

if total_belanja >= 150_000 and estimasi_jam < 22:
    if metode_bayar in ("ovo", "gopay"):
        print("Promo cashback 20 ribu + gratis ongkir")
    else:
        print("Gratis ongkir tanpa cashback")
elif total_belanja >= 100_000:
    print("Gratis ongkir hingga 10 ribu")
else:
    print("Belanja lagi untuk menikmati promo")
````
