# Python Functions

Ringkasan ini mengadaptasi konten GeeksforGeeks tentang fungsi Python, mencakup definisi, jenis argumen, hingga fungsi rekursif.

Fungsi adalah sekumpulan pernyataan yang menjalankan tugas khusus. Dengan membungkus logika berulang menjadi fungsi, kita bisa memanggilnya berkali-kali untuk berbagai input tanpa menyalin kode.

## Mengapa Menggunakan Fungsi?

- **Code Reusability**: logika dapat dipanggil ulang.
- **Modularity**: program besar dipecah menjadi bagian kecil.
- **Readability**: struktur program lebih mudah dibaca.
- **Maintainability**: perubahan cukup dilakukan di satu tempat.

## Mendefinisikan Fungsi

Gunakan kata kunci `def` diikuti nama fungsi dan parameter opsional:

- $f(x) = x+2$
- $f(5) = x+2 = 7$

```python
def function_name(parameters):
    """Docstring opsional"""
    # isi fungsi
    return ekspresi
```

````python
# contoh fungsi dasar
def fun(x: str):
    print(x)

fun("Hello SMKN 6 Makassar")
````

Memanggil fungsi dilakukan dengan menuliskan namanya diikuti tanda kurung dan argumen yang diperlukan.

## Fungsi dengan Nilai Kembali
Contoh berikut mengembalikan "Even" atau "Odd" bergantung parameter.

````python
# fungsi yang mengembalikan nilai
def even_odd(x):
    if x % 2 == 0:#habis dibagi dengan 2
        return "Even"#kembalikan nilai genap
    return "Odd"#ganjil

print(even_odd(678909876789))
print(even_odd(87348953839586))
````

## Jenis Argumen Fungsi
GeeksforGeeks mengelompokkan argumen menjadi beberapa tipe: default, keyword, positional, dan arbitrary (`*args`, `**kwargs`).

### 1. Default Arguments
Memberikan nilai bawaan jika pemanggil tidak menyuplai nilai tertentu.

````python
for i in range(0, 100, 1):
    print(i)

for i in range(100):
    print(i)
````

````python
# default argument
def show_point(x: int, y=50):
    print("x:", x)
    print("y:", y)

show_point("Kambing")# x:10, y:50
show_point(10, 20)# x:10, y:20
````

### 2. Keyword Arguments
Argumen disuplai dengan menyebut nama parameternya sehingga urutan tidak penting.

````python
# keyword arguments
def student(fname, lname):
    print(fname, lname)

student(fname='Geeks', lname='Practice')
student(lname='Practice', fname='Geeks')
````

### 3. Positional Arguments
Nilai dipetakan sesuai urutan parameter yang didefinisikan.

````python
# positional arguments
def name_age(name, age):
    print("Hi, I am", name)
    print("My age is", age)

print("Case-1:")
name_age("Suraj", 27)

print("Case-2:")
name_age(27, "Suraj")
````

### 4. Arbitrary Arguments (`*args` dan `**kwargs`)
Mengizinkan jumlah argumen yang bervariasi, baik tanpa nama maupun pasangan kunci-nilai.

````python
def function(x, y):
    print(x,y)

function(6,9,10)
````

````python
# arbitrary arguments
def my_fun(*args, **kwargs):
    print("Non-keyword args:")
    for arg in args:
        print(arg)
    print("Keyword args:")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

my_fun('Hey', 'Welcome', 'to', 'SMKN 6 Makassar', first='Geeks', mid='for', last='Geeks')
````

## Statement `return`
`return` mengakhiri eksekusi fungsi dan mengirimkan nilai kembali ke pemanggil. Dapat mengembalikan tipe apa pun.

````python
# contoh return statement
def square_value(num):
    """Mengembalikan kuadrat dari input"""
    return num ** 2

print(square_value(2))
print(square_value(-4))
````

````python
# contoh return statement
def square_value(num):
    """Mengembalikan kuadrat dari input"""
    square = num ** 3
    return square

print(square_value(2))
print(square_value(-4))
````

## Pass by Reference vs Pass by Value
Python memakai konsep *pass-by-object-reference*: objek mutable dapat dimodifikasi dari dalam fungsi, sementara objek immutable tidak berubah.

````python
# efek pada objek mutable vs immutable
def mutate_list(lst):
    lst[0] = 20

nums = [10, 11, 12, 13]
mutate_list(nums)
print(nums)  # list berubah

def mutate_number(val):
    val = 20

n = 10
mutate_number(n)
print(n)  # integer tetap
````

## Function within Functions
Fungsi yang dideklarasikan di dalam fungsi lain disebut inner function, berguna untuk menjaga scope tetap terkontrol.

````python
# inner function
def outer():
    s = 'I love learn math'
    def inner():
        print(s)
    inner()

outer()
````

## Anonymous Functions (Lambda)
`lambda` membuat fungsi tanpa nama untuk operasi sederhana.

````python
# lambda vs fungsi biasa
def cube(s):
    return s * s * s

cube_l = lambda x: x * x * x

print(cube(7))
print(cube_l(7))
````

## Recursive Functions
Fungsi rekursif memanggil dirinya sendiri untuk memecahkan masalah yang lebih kecil. Selalu definisikan *base case* agar tidak terjadi rekursi tak berujung.

- $ 3! = 1*2*3 = 6$
- $ 4! = 1*2*3*4 = 24$
- $ 4! = 4*3*2*1 = 24$

````python
# contoh rekursi: faktorial
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

#5
#x
#4
#x
#3



print(factorial(4))
````
