```python
# Fungsi untuk menghitung luas bangun datar
def hitung_luas_persegi(sisi):
    return sisi ** 2

def hitung_luas_lingkaran(radius):
    return 3.141592653589793 * radius ** 2  # Menggunakan nilai π secara manual

def hitung_luas_segitiga(alas, tinggi):
    return 0.5 * alas * tinggi

# History list to keep track of all calculations
history = []

def print_history():
    if not history:
        print("Belum ada riwayat perhitungan.")
    else:
        print("Riwayat Perhitungan:")
        for i, (shape, area) in enumerate(history, 1):
            shape_name = {
                '1': 'Persegi',
                '2': 'Lingkaran',
                '3': 'Segitiga'
            }.get(shape, 'Unknown')
            print(f"{i}. {shape_name} - Luas: {area} satuan kuadrat")

# Kalkulator utama
def kalkulator():
    print("Selamat datang di Kalkulator Luas Bangun Datar!")

    while True:
        print("\nPilih bangun datar yang ingin dihitung luasnya:")
        print("1. Persegi")
        print("2. Lingkaran")
        print("3. Segitiga")
        print("4. Keluar")
        print("5. History")
        
        pilihan = input("Masukkan pilihan (1/2/3/4): ").strip()
        
        if pilihan == '4':
            print("Terima kasih telah menggunakan kalkulator!")
            break
        
        elif pilihan == '1':  # Persegi
            sisi = float(input("Masukkan panjang sisi persegi: "))
            luas = hitung_luas_persegi(sisi)
            print(f"Luas persegi adalah: {luas} satuan kuadrat")
        
        elif pilihan == '2':  # Lingkaran
            radius = float(input("Masukkan radius lingkaran: "))
            luas = hitung_luas_lingkaran(radius)
            print(f"Luas lingkaran adalah: {luas} satuan kuadrat")
        
        elif pilihan == '3':  # Segitiga
            alas = float(input("Masukkan panjang alas segitiga: "))
            tinggi = float(input("Masukkan tinggi segitiga: "))
            luas = hitung_luas_segitiga(alas, tinggi)
            print(f"Luas segitiga adalah: {luas} satuan kuadrat")
        
        elif pilihan == '5':  # History
            print_history()
        
        else:
            print("Pilihan tidak valid, silakan pilih lagi.")

        history.append((pilihan, luas))

if __name__ == "__main__":
    kalkulator()
```

```python
import tkinter as tk

# --------------------
#  FUNGSI LOGIKA
# --------------------
expression = ""        # dipakai untuk perhitungan (pakai *, /, dll)
display_text = ""      # dipakai untuk ditampilkan (×, ÷, dll)

def update_display():
    """Update teks yang muncul di layar."""
    display_var.set(display_text if display_text else "0")

def add_to_expression(value_for_calc, value_for_display=None):
    """Tambahkan angka / operator ke ekspresi."""
    global expression, display_text

    if value_for_display is None:
        value_for_display = value_for_calc

    expression += str(value_for_calc)
    display_text += str(value_for_display)
    update_display()

def clear_all():
    """Tombol C: hapus semua."""
    global expression, display_text
    expression = ""
    display_text = ""
    update_display()

def backspace():
    """Tombol ← : hapus satu karakter terakhir."""
    global expression, display_text
    if expression:
        expression = expression[:-1]
        display_text = display_text[:-1]
        update_display()

def calculate():
    """Tombol = : hitung hasil."""
    global expression, display_text
    if not expression:
        return
    try:
        result = eval(expression)
        expression = str(result)
        display_text = expression
        update_display()
    except Exception:
        expression = ""
        display_text = "Error"
        update_display()

def percentage():
    """Tombol % : anggap ekspresi sebagai angka lalu /100."""
    global expression, display_text
    if not expression:
        return
    try:
        result = eval(expression) / 100
        expression = str(result)
        display_text = expression
        update_display()
    except Exception:
        expression = ""
        display_text = "Error"
        update_display()

# --------------------
#  GUI TKINTER
# --------------------
root = tk.Tk()
root.title("Calculator")
root.geometry("320x480")
root.configure(bg="#0f172a")  # background biru gelap

# Frame utama (kartu kalkulator)
card = tk.Frame(root, bg="#f9fafb", bd=0, padx=16, pady=16)
card.place(relx=0.5, rely=0.5, anchor="center")

# Judul
title_label = tk.Label(
    card,
    text="Calculator",
    font=("Segoe UI", 12, "bold"),
    bg="#f9fafb",
    fg="#374151"
)
title_label.pack(anchor="n", pady=(0, 8))

# Layar hitam
display_frame = tk.Frame(card, bg="#020617", height=80)
display_frame.pack(fill="x", pady=(0, 12))

display_var = tk.StringVar()
display_label = tk.Label(
    display_frame,
    textvariable=display_var,
    font=("Segoe UI", 32, "bold"),
    bg="#020617",
    fg="#f9fafb",
    anchor="e",
    padx=10,
    pady=10
)
display_label.pack(fill="both")

update_display()  # set awal 0

# Frame tombol
btn_frame = tk.Frame(card, bg="#f9fafb")
btn_frame.pack()

# Style tombol
btn_font = ("Segoe UI", 14, "bold")

def create_btn(text, row, col, width=4, height=2,
               bg="#e5e7eb", fg="#111827",
               command=None, col_span=1):
    btn = tk.Button(
        btn_frame,
        text=text,
        font=btn_font,
        bg=bg,
        fg=fg,
        activebackground=bg,
        activeforeground=fg,
        bd=0,
        relief="flat",
        width=width,
        height=height,
        command=command
    )
    btn.grid(row=row, column=col, columnspan=col_span, padx=4, pady=4, sticky="nsew")
    return btn

# Atur grid agar responsif
for i in range(4):
    btn_frame.columnconfigure(i, weight=1)
for i in range(5):
    btn_frame.rowconfigure(i, weight=1)

# Baris 1: C   %   ←   +
create_btn("C", 0, 0, command=clear_all, bg="#e5e7eb")
create_btn("%", 0, 1, command=percentage, bg="#e5e7eb")
create_btn("←", 0, 2, command=backspace, bg="#e5e7eb")
create_btn("+", 0, 3,
           command=lambda: add_to_expression("+"),
           bg="#f97316", fg="white")

# Baris 2: 7 8 9 ×
create_btn("7", 1, 0, command=lambda: add_to_expression("7"))
create_btn("8", 1, 1, command=lambda: add_to_expression("8"))
create_btn("9", 1, 2, command=lambda: add_to_expression("9"))
create_btn("×", 1, 3,
           command=lambda: add_to_expression("*", "×"),
           bg="#f97316", fg="white")

# Baris 3: 4 5 6 -
create_btn("4", 2, 0, command=lambda: add_to_expression("4"))
create_btn("5", 2, 1, command=lambda: add_to_expression("5"))
create_btn("6", 2, 2, command=lambda: add_to_expression("6"))
create_btn("-", 2, 3,
           command=lambda: add_to_expression("-"),
           bg="#f97316", fg="white")

# Baris 4: 1 2 3 +
create_btn("1", 3, 0, command=lambda: add_to_expression("1"))
create_btn("2", 3, 1, command=lambda: add_to_expression("2"))
create_btn("3", 3, 2, command=lambda: add_to_expression("3"))
create_btn("÷", 3, 3,
           command=lambda: add_to_expression("/", "÷"),
           bg="#f97316", fg="white")

# Baris 5: 0 (span 3 kolom)  =
create_btn("0", 4, 0, col_span=3,
           command=lambda: add_to_expression("0"))
create_btn("=", 4, 3,
           command=calculate,
           bg="#f97316", fg="white")

root.mainloop()
```
