```python
import tkinter as tk
from tkinter import ttk

root = tk.Tk()
root.title("Checkbutton Example")

def on_check():
    if agree_var.get():
        print("User agreed to the terms.")
    else:
        print("User did not agree to the terms.")

# State variable
agree_var = tk.BooleanVar()

check = ttk.Checkbutton(
    root, 
    text="Saya setuju dengan syarat dan ketentuan", 
    variable=agree_var)

button = ttk.Button(
    root,
    text="Submit",
    command=on_check
)
button.pack()

check.pack(padx=20, pady=20)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
import csv
import os

# ========= CSV SAVER FUNCTION =========
def save_to_csv(data, filename="biodata.csv"):
    file_exists = os.path.exists(filename)

    with open(filename, mode="a", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)

        # Tulis header kalau file baru
        if not file_exists:
            writer.writerow([
                "Name", "Age", "Gender", "Phone",
                "Email", "City", "Hobbies", "About"
            ])

        writer.writerow([
            data["name"],
            data["age"],
            data["gender"],
            data["phone"],
            data["email"],
            data["city"],
            ", ".join(data["hobbies"]) if data["hobbies"] else "",
            data["about"],
        ])

    print(f"Saved to {filename}")


# ========= TKINTER UI =========
root = tk.Tk()
root.title("Biodata Form")
root.geometry("600x400")

# ---- main container ----
main_frame = ttk.Frame(root, padding=10)
main_frame.pack(fill="both", expand=True)

# 2-column layout frames
left_frame = ttk.Frame(main_frame)
right_frame = ttk.Frame(main_frame)

left_frame.grid(row=0, column=0, sticky="nwe", padx=(0, 10))
right_frame.grid(row=0, column=1, sticky="nwe")

main_frame.columnconfigure(0, weight=1)
main_frame.columnconfigure(1, weight=1)

# ===== STATE VARIABLES =====
name_var = tk.StringVar()
age_var = tk.IntVar(value=18)
gender_var = tk.StringVar(value="Not specified")
phone_var = tk.StringVar()
email_var = tk.StringVar()
city_var = tk.StringVar()

hobby_music = tk.BooleanVar()
hobby_sport = tk.BooleanVar()
hobby_read = tk.BooleanVar()

# ===== LEFT SIDE =====
ttk.Label(left_frame, text="Name:").grid(row=0, column=0, sticky="w")
ttk.Entry(left_frame, textvariable=name_var, width=25).grid(row=0, column=1, sticky="we", pady=2)

ttk.Label(left_frame, text="Age:").grid(row=1, column=0, sticky="w")
age_spin = ttk.Spinbox(left_frame, from_=1, to=120, textvariable=age_var, width=5)
age_spin.grid(row=1, column=1, sticky="w", pady=2)

ttk.Label(left_frame, text="Phone:").grid(row=2, column=0, sticky="w")
ttk.Entry(left_frame, textvariable=phone_var, width=25).grid(row=2, column=1, sticky="we", pady=2)

ttk.Label(left_frame, text="Email:").grid(row=3, column=0, sticky="w")
ttk.Entry(left_frame, textvariable=email_var, width=25).grid(row=3, column=1, sticky="we", pady=2)

left_frame.columnconfigure(1, weight=1)

# ===== RIGHT SIDE =====
ttk.Label(right_frame, text="Gender:").grid(row=0, column=0, sticky="w")

gender_frame = ttk.Frame(right_frame)
gender_frame.grid(row=0, column=1, sticky="w")

ttk.Radiobutton(gender_frame, text="Male",   value="Male",   variable=gender_var).pack(side="left")
ttk.Radiobutton(gender_frame, text="Female", value="Female", variable=gender_var).pack(side="left")
ttk.Radiobutton(gender_frame, text="Other",  value="Other",  variable=gender_var).pack(side="left")

ttk.Label(right_frame, text="City:").grid(row=1, column=0, sticky="w")
city_combo = ttk.Combobox(right_frame, textvariable=city_var, state="readonly", width=22)
city_combo["values"] = ("Makassar", "Palu", "Jakarta", "Surabaya", "Other")
city_combo.grid(row=1, column=1, sticky="w", pady=2)
city_combo.current(0)

ttk.Label(right_frame, text="Hobbies:").grid(row=2, column=0, sticky="nw")
hobby_frame = ttk.Frame(right_frame)
hobby_frame.grid(row=2, column=1, sticky="w")

ttk.Checkbutton(hobby_frame, text="Music",   variable=hobby_music).pack(anchor="w")
ttk.Checkbutton(hobby_frame, text="Sport",   variable=hobby_sport).pack(anchor="w")
ttk.Checkbutton(hobby_frame, text="Reading", variable=hobby_read).pack(anchor="w")

ttk.Label(right_frame, text="About you:").grid(row=3, column=0, sticky="nw", pady=(5, 0))
about_text = tk.Text(right_frame, width=30, height=5)
about_text.grid(row=3, column=1, sticky="we", pady=(5, 0))

right_frame.columnconfigure(1, weight=1)

# ===== SUBMIT BUTTON =====
def on_submit():
    hobbies = []
    if hobby_music.get():
        hobbies.append("Music")
    if hobby_sport.get():
        hobbies.append("Sport")
    if hobby_read.get():
        hobbies.append("Reading")

    about = about_text.get("1.0", "end-1c")

    data = {
        "name": name_var.get(),
        "age": age_var.get(),
        "gender": gender_var.get(),
        "phone": phone_var.get(),
        "email": email_var.get(),
        "city": city_var.get(),
        "hobbies": hobbies,
        "about": about,
    }

    save_to_csv(data)  # ‚Üê sekarang simpan ke CSV

submit_btn = ttk.Button(main_frame, text="Save to CSV", command=on_submit)
submit_btn.grid(row=1, column=0, columnspan=2, pady=10)

root.mainloop()
```

```python
import tkinter as tk

def aksi_tombol(nomor):
    print(f"Tombol {nomor} diklik!")

jendela = tk.Tk()
jendela.title("Layout 3x3 Grid")

# Buat 9 tombol
tombol = []
for i in range(1, 10):
    # Membuat Tombol
    btn = tk.Button(jendela, 
                    text=str(i), 
                    width=10, 
                    height=3,
                    # Menggunakan lambda untuk meneruskan parameter ke fungsi aksi_tombol
                    command=lambda n=i: aksi_tombol(n))
    tombol.append(btn)
    
# Mengatur layout menggunakan grid()
# Kita ingin 3 baris dan 3 kolom (total 9 tombol)
row = 0
col = 0
for i in range(9):
    # Tombol saat ini
    btn = tombol[i]
    
    # 1. Menentukan posisi grid
    # btn.grid(row=baris, column=kolom)
    btn.grid(row=row, column=col, padx=5, pady=5) # padx/pady memberi jarak antar tombol
    
    # 2. Logika untuk memindahkan posisi (pindah kolom, jika kolom penuh, pindah baris)
    col += 1
    if col > 2: # Jika sudah mengisi kolom 0, 1, dan 2
        col = 0 # Reset ke kolom pertama
        row += 1 # Pindah ke baris berikutnya

# Menambahkan konfigurasi agar grid menyesuaikan ukuran jendela
# Ini penting agar tombol tetap proporsional saat jendela diubah ukurannya
for i in range(3):
    jendela.grid_columnconfigure(i, weight=1)
    jendela.grid_rowconfigure(i, weight=1)

jendela.mainloop()
```

## 5. Mengatur Layout: `pack` vs `grid`
Tkinter menyediakan beberapa manajer layout. `pack` cocok untuk susunan vertikal/horizontal sederhana, sedangkan `grid` kuat untuk tata letak tabel. Kamu bisa mengombinasikannya dengan membuat `Frame` untuk setiap area.

```python
import tkinter as tk

root = tk.Tk()
root.title("Demo Layout")
root.geometry("360x240")

header = tk.Frame(root, bg="#3d405b")
header.pack(fill="x")

title = tk.Label(header, text="Form Kontak", fg="#fff", bg="#3d405b", font=("Helvetica", 16, "bold"))
title.pack(pady=8)

content = tk.Frame(root)
content.pack(fill="both", expand=True, padx=16, pady=12)

# Menggunakan grid di dalam frame content
tk.Label(content, text="Nama:").grid(row=0, column=0, sticky="w", padx=4, pady=4)
entry_nama = tk.Entry(content, width=25)
entry_nama.grid(row=0, column=1, padx=4, pady=4)

tk.Label(content, text="Email:").grid(row=1, column=0, sticky="w", padx=4, pady=4)
entry_email = tk.Entry(content, width=25)
entry_email.grid(row=1, column=1, padx=4, pady=4)

kirim_btn = tk.Button(content, text="Simpan", bg="#81b29a", fg="#fff")
kirim_btn.grid(row=2, column=0, columnspan=2, pady=(10, 0))

content.columnconfigure(1, weight=1)

root.mainloop()
```

## 8. Proyek Akhir: Kalkulator Sederhana
Sekarang kita gabungkan semua konsep sebelumnya untuk membuat aplikasi kalkulator sederhana. Fungsinya:
- Menampilkan layar input (Entry) yang memperlihatkan ekspresi saat tombol ditekan.
- Tombol angka dan operator disusun memakai `grid`.
- Tombol `C` untuk menghapus dan `=` untuk menghitung dengan `eval` (dibungkus `try/except`).

Perhatikan bagaimana kita menggunakan `Frame`, `ttk.Style`, fungsi callback, serta `StringVar` untuk menyimpan ekspresi.

```python
import tkinter as tk
from tkinter import ttk

root = tk.Tk()
root.title("Kalkulator Sederhana")
root.geometry("300x380")
root.resizable(False, False)

style = ttk.Style()
style.configure("Display.TEntry", font=("Helvetica", 20))
style.configure("Key.TButton", font=("Helvetica", 14), padding=6)

expression_var = tk.StringVar()


def tambah_input(value: str):
    expression_var.set(expression_var.get() + value)


def bersihkan():
    expression_var.set("")


def hitung_hasil():
    ekspresi = expression_var.get()
    try:
        hasil = eval(ekspresi)
        expression_var.set(str(hasil))
    except Exception:
        expression_var.set("Error")


display = ttk.Entry(root, textvariable=expression_var, justify="right", style="Display.TEntry")
display.pack(fill="x", padx=16, pady=(20, 10))

grid_frame = ttk.Frame(root, padding=10)
grid_frame.pack(fill="both", expand=True)

buttons = [
    ["7", "8", "9", "/"],
    ["4", "5", "6", "*"],
    ["1", "2", "3", "-"],
    ["0", ".", "=", "+"],
]

for r, row in enumerate(buttons):
    for c, char in enumerate(row):
        if char == "=":
            btn = ttk.Button(grid_frame, text=char, style="Key.TButton", command=hitung_hasil)
        else:
            btn = ttk.Button(grid_frame, text=char, style="Key.TButton", command=lambda ch=char: tambah_input(ch))
        btn.grid(row=r + 1, column=c, sticky="nsew", padx=4, pady=4)

clear_btn = ttk.Button(grid_frame, text="C", style="Key.TButton", command=bersihkan)
clear_btn.grid(row=0, column=0, columnspan=4, sticky="nsew", padx=4, pady=4)

for i in range(4):
    grid_frame.columnconfigure(i, weight=1)
for j in range(5):
    grid_frame.rowconfigure(j, weight=1)

root.mainloop()
```

Selanjutnya kamu bisa bereksperimen dengan widget lain seperti `Listbox`, `Canvas`, `Menu`, hingga membuat aplikasi yang lebih kompleks.

```python
import tkinter as tk
from datetime import datetime

root = tk.Tk()
root.title("My To-Do List")
root.geometry("420x520")
root.config(bg="#FEF3E2")  # soft cream

# ==== FONTS ====
TITLE_FONT = ("Helvetica", 20, "bold")
DATETIME_FONT = ("Helvetica", 10)
ENTRY_FONT = ("Helvetica", 12)
LIST_FONT = ("Helvetica", 11)

# ==== TITLE ====
title_label = tk.Label(
    root,
    text="To-Do List",
    font=TITLE_FONT,
    bg="#FEF3E2",
    fg="#FA812F"   # bright orange
)
title_label.pack(pady=(10, 2))

# ==== INFO / CURRENT DATE (optional) ====
info_label = tk.Label(
    root,
    text="Set the date & time for your task ‚Üì",
    font=DATETIME_FONT,
    bg="#FEF3E2",
    fg="#DD0303"
)
info_label.pack(pady=(0, 8))

# ==== DATE & TIME PICKER FRAME ====
dt_frame = tk.Frame(root, bg="#FEF3E2")
dt_frame.pack(padx=10, fill="x")

# Date
today_str = datetime.now().strftime("%d-%m-%Y")
date_var = tk.StringVar(value=today_str)

lbl_date = tk.Label(dt_frame, text="Date (DD-MM-YYYY):", bg="#FEF3E2")
lbl_date.grid(row=0, column=0, sticky="w")

entry_date = tk.Entry(
    dt_frame,
    textvariable=date_var,
    font=("Helvetica", 10),
    bg="#FAB12F",
    fg="black",
    relief="flat"
)
entry_date.grid(row=0, column=1, sticky="we", padx=(5, 0))

# Time (Hour & Minute)
lbl_time = tk.Label(dt_frame, text="Time (HH:MM):", bg="#FEF3E2")
lbl_time.grid(row=1, column=0, sticky="w", pady=(5, 0))

time_frame = tk.Frame(dt_frame, bg="#FEF3E2")
time_frame.grid(row=1, column=1, sticky="w", pady=(5, 0))

hour_var = tk.StringVar(value="08")
minute_var = tk.StringVar(value="00")

spin_hour = tk.Spinbox(
    time_frame,
    from_=0, to=23,
    width=3,
    textvariable=hour_var,
    font=("Helvetica", 10),
    bg="#FAB12F",
    fg="black",
    relief="flat",
    justify="center",
    format="%02.0f"
)
spin_hour.pack(side="left")

lbl_colon = tk.Label(time_frame, text=":", bg="#FEF3E2")
lbl_colon.pack(side="left")

spin_minute = tk.Spinbox(
    time_frame,
    from_=0, to=59,
    width=3,
    textvariable=minute_var,
    font=("Helvetica", 10),
    bg="#FAB12F",
    fg="black",
    relief="flat",
    justify="center",
    format="%02.0f"
)
spin_minute.pack(side="left", padx=(2, 0))

dt_frame.columnconfigure(1, weight=1)

# ==== ENTRY + ADD BUTTON ====
input_frame = tk.Frame(root, bg="#FEF3E2")
input_frame.pack(pady=10, padx=10, fill="x")

task_var = tk.StringVar()

entry_task = tk.Entry(
    input_frame,
    textvariable=task_var,
    font=ENTRY_FONT,
    bg="#FAB12F",      # warm gold
    fg="#000000",
    insertbackground="#000000",
    relief="flat"
)
entry_task.pack(side="left", fill="x", expand=True, padx=(0, 8))


def add_task():
    text = task_var.get().strip()
    date_text = date_var.get().strip()
    hour_text = hour_var.get().strip()
    minute_text = minute_var.get().strip()

    if not text:
        return  # nothing to add

    # sangat basic: tidak validasi format tanggal/jam dulu, hanya gabungkan
    # kalau mau, bisa ditambah try/except datetime.strptime untuk validasi
    time_str = f"{hour_text.zfill(2)}:{minute_text.zfill(2)}"
    stamp = f"{date_text} {time_str}"
    item_text = f"[{stamp}] {text}"
    listbox_tasks.insert(tk.END, item_text)
    task_var.set("")


btn_add = tk.Button(
    input_frame,
    text="Add",
    font=("Helvetica", 11, "bold"),
    bg="#FA812F",
    fg="black",
    activebackground="#FAB12F",
    activeforeground="black",
    relief="flat",
    command=add_task
)
btn_add.pack(side="right")

# ==== LISTBOX + SCROLLBAR ====
list_frame = tk.Frame(root, bg="#FEF3E2")
list_frame.pack(padx=10, pady=(5, 10), fill="both", expand=True)

scrollbar = tk.Scrollbar(list_frame)
scrollbar.pack(side="right", fill="y")

listbox_tasks = tk.Listbox(
    list_frame,
    font=LIST_FONT,
    bg="#FA812F",
    fg="#000000",
    selectbackground="#FAB12F",
    selectforeground="#000000",
    activestyle="none",
    yscrollcommand=scrollbar.set
)
listbox_tasks.pack(fill="both", expand=True)

scrollbar.config(command=listbox_tasks.yview)

# ==== BUTTONS: DELETE & CLEAR ====
buttons_frame = tk.Frame(root, bg="#FEF3E2")
buttons_frame.pack(pady=10)

def delete_selected():
    selection = listbox_tasks.curselection()
    if selection:
        listbox_tasks.delete(selection[0])

def clear_all():
    listbox_tasks.delete(0, tk.END)

btn_delete = tk.Button(
    buttons_frame,
    text="Delete Selected",
    font=("Helvetica", 11),
    bg="#DD0303",   # red
    fg="black",
    activebackground="#B50202",
    activeforeground="black",
    relief="flat",
    command=delete_selected
)
btn_delete.pack(side="left", padx=5)

btn_clear = tk.Button(
    buttons_frame,
    text="Clear All",
    font=("Helvetica", 11),
    bg="#FAB12F",
    fg="black",
    activebackground="#FA812F",
    activeforeground="black",
    relief="flat",
    command=clear_all
)
btn_clear.pack(side="left", padx=5)

# Optional: Enter key adds task
root.bind("<Return>", lambda event: add_task())

root.mainloop()
```
