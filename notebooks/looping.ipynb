# Loops in Python (For, While, dan Nested)

Ringkasan ini diadaptasi dari materi GeeksforGeeks tentang perulangan. Fokusnya adalah bagaimana `for`, `while`, dan perulangan bersarang bekerja, beserta kontrol `break`, `continue`, dan `pass`.

Perulangan memungkinkan blok kode dijalankan berkali-kali sehingga tugas yang repetitif tidak perlu ditulis manual. Python menyediakan `for` loop yang cocok untuk mengiterasi koleksi, `while` loop untuk proses berbasis kondisi, serta mekanisme kontrol yang menjaga alur tetap aman.

## For Loop

`for` dipakai untuk mengiterasi setiap elemen pada sequence (list, tuple, string) atau objek iterable lainnya. Bentuk umumnya `for variabel in iterable:` lalu blok kode dijalankan untuk tiap elemen. Contoh pada materi menunjukkan perulangan dari 0 sampai kurang dari `n`.

```python
# contoh 1: for loop dengan range
n = 5
for i in range(0, n):
    print(i)
```

### Iterasi Berdasarkan Indeks Sequence

Materi juga menyoroti teknik mengakses elemen lewat indeks menggunakan `range(len(koleksi))`. Ini berguna saat kita ingin mengetahui posisi dan nilai sekaligus.

```python
# contoh 2: iterasi list berdasarkan indeks
arr = ["Geeks", "For", "Geeks"]
for idx in range(len(arr)):
    print(f"Elemen ke-{idx}: {arr[idx]}")
```

## While Loop

`while` menjalankan blok kode selama kondisi bernilai `True`. Materi pada gambar menekankan bahwa kita harus memperbarui variabel kondisi agar loop tidak berjalan tanpa akhir.

```python
# contoh 3: while loop sederhana
count = 0
while count < 3:
    count += 1
    print("Hello Geek")
```

## Nested Loop (Perulangan Bersarang)

Materi memperlihatkan nested loop untuk mengunjungi kombinasi dua indeks. Loop luar berpindah baris, loop dalam memproses setiap kolom atau detail lain.

```python
# contoh 5: nested loop
for i in range(1, 4):
    for j in range(1, 3):
        print(f"Pasangan (i={i}, j={j})")
```

## Loop Control Statements

Bagian akhir materi membahas `break`, `continue`, dan `pass` untuk mengendalikan alur eksekusi di dalam loop.

### Break
Memaksa loop berhenti sepenuhnya ketika syarat tertentu terpenuhi.

```python
# contoh 6: penggunaan break
for angka in range(10):
    if angka == 6:
        print("berhenti di", angka)
        break
    print(angka)
```

```python
# contoh: menghentikan while menggunakan break
i = 0
while True: #sampai tak hingga
    i += 1
    print("Iterasi ke", i)
    if i == 990000000:
        print("Angka 990000000 ditemukan, menghentikan loop dengan break.")
        break
else:
    # blok else tidak akan dijalankan karena loop dihentikan dengan break
    print("Loop selesai tanpa break")
```

### Continue
Melewati iterasi saat ini dan melanjutkan ke iterasi berikutnya tanpa keluar dari loop.

```python
# contoh 7: penggunaan continue
for angka in range(10):
    if angka % 2 == 0:
        continue
    print("bilangan ganjil:", angka)
```

```python
# contoh: penggunaan continue di dalam while
# Cetak bilangan ganjil di bawah 10. Penting: pastikan variabel penghitung diperbarui
# sebelum `continue` untuk menghindari infinite loop.

n = 10
i = 0
while i < n:
    i += 1                # update penghitung selalu sebelum menggunakan continue
    if i % 2 == 0:
        # lewati bilangan genap, lanjut ke iterasi berikutnya
        continue
    print("bilangan ganjil:", i)

# Catatan singkat (contoh salah, jangan jalankan):
# Jika increment ditempatkan setelah continue, loop bisa menjadi infinite loop.
# i = 0
# while i < 3:
#     if i == 1:
#         continue  # i tidak bertambah di sini -> infinite loop
#     i += 1
```

### Pass
`pass` bertindak sebagai placeholder ketika blok kode harus ada secara sintaks, tetapi belum ingin diisi logika.

```python
# contoh 8: pass di dalam loop
for huruf in "geeksforgeeks":
    pass
print("Loop selesai dengan huruf terakhir:", huruf)
```

```python
# contoh: menggunakan `pass` di dalam while sebagai placeholder
# Pastikan variabel kondisi selalu diperbarui agar tidak menyebabkan infinite loop.

i = 0
while i < 5:
    i += 1
    if i == 3:
        # belum ingin mengisi logika khusus untuk i == 3,
        # gunakan `pass` sebagai placeholder agar sintaks valid
        pass
    print("nilai i =", i)

# Contoh yang salah (JANGAN JALANKAN) — menyebabkan infinite loop karena kondisi tidak diperbarui:
# i = 0
# while i < 2:
#     if i == 0:
#         pass   # tanpa update i di tempat lain, loop tidak akan berakhir
#     # i += 1  # jika baris ini dikomentari, loop akan infinite
```

## Catatan Penting dari Materi
- Pastikan variabel penghitung diperbarui agar `while` tidak menjadi infinite loop.
- Gunakan `range(start, stop, step)` untuk mengontrol iterasi `for`.
- Nested loop cepat membengkak jumlah iterasinya, jadi rencanakan batasnya dengan baik.
