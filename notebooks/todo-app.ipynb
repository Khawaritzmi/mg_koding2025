```python
import tkinter as tk
from datetime import datetime

root = tk.Tk()
root.title("My To-Do List")
root.geometry("420x520")
root.config(bg="#FEF3E2")  # soft cream

# ==== FONTS ====
TITLE_FONT = ("Helvetica", 20, "bold")
DATETIME_FONT = ("Helvetica", 10)
ENTRY_FONT = ("Helvetica", 12)
LIST_FONT = ("Helvetica", 11)

# ==== TITLE ====
title_label = tk.Label(
    root,
    text="To-Do List",
    font=TITLE_FONT,
    bg="#FEF3E2",
    fg="#FA812F"   # bright orange
)
title_label.pack(pady=(10, 2))

# ==== INFO / CURRENT DATE (optional) ====
info_label = tk.Label(
    root,
    text="Set the date & time for your task â†“",
    font=DATETIME_FONT,
    bg="#FEF3E2",
    fg="#DD0303"
)
info_label.pack(pady=(0, 8))

# ==== DATE & TIME PICKER FRAME ====
dt_frame = tk.Frame(root, bg="#FEF3E2")
dt_frame.pack(padx=10, fill="x")

# Date
today_str = datetime.now().strftime("%d-%m-%Y")
date_var = tk.StringVar(value=today_str)

lbl_date = tk.Label(dt_frame, text="Date (DD-MM-YYYY):", bg="#FEF3E2")
lbl_date.grid(row=0, column=0, sticky="w")

entry_date = tk.Entry(
    dt_frame,
    textvariable=date_var,
    font=("Helvetica", 10),
    bg="#FAB12F",
    fg="black",
    relief="flat"
)
entry_date.grid(row=0, column=1, sticky="we", padx=(5, 0))

# Time (Hour & Minute)
lbl_time = tk.Label(dt_frame, text="Time (HH:MM):", bg="#FEF3E2")
lbl_time.grid(row=1, column=0, sticky="w", pady=(5, 0))

time_frame = tk.Frame(dt_frame, bg="#FEF3E2")
time_frame.grid(row=1, column=1, sticky="w", pady=(5, 0))

hour_var = tk.StringVar(value="08")
minute_var = tk.StringVar(value="00")

spin_hour = tk.Spinbox(
    time_frame,
    from_=0, to=23,
    width=3,
    textvariable=hour_var,
    font=("Helvetica", 10),
    bg="#FAB12F",
    fg="black",
    relief="flat",
    justify="center",
    format="%02.0f"
)
spin_hour.pack(side="left")

lbl_colon = tk.Label(time_frame, text=":", bg="#FEF3E2")
lbl_colon.pack(side="left")

spin_minute = tk.Spinbox(
    time_frame,
    from_=0, to=59,
    width=3,
    textvariable=minute_var,
    font=("Helvetica", 10),
    bg="#FAB12F",
    fg="black",
    relief="flat",
    justify="center",
    format="%02.0f"
)
spin_minute.pack(side="left", padx=(2, 0))

dt_frame.columnconfigure(1, weight=1)

# ==== ENTRY + ADD BUTTON ====
input_frame = tk.Frame(root, bg="#FEF3E2")
input_frame.pack(pady=10, padx=10, fill="x")

task_var = tk.StringVar()

entry_task = tk.Entry(
    input_frame,
    textvariable=task_var,
    font=ENTRY_FONT,
    bg="#FAB12F",      # warm gold
    fg="#000000",
    insertbackground="#000000",
    relief="flat"
)
entry_task.pack(side="left", fill="x", expand=True, padx=(0, 8))


def add_task():
    text = task_var.get().strip()
    date_text = date_var.get().strip()
    hour_text = hour_var.get().strip()
    minute_text = minute_var.get().strip()

    if not text:
        return  # nothing to add

    # sangat basic: tidak validasi format tanggal/jam dulu, hanya gabungkan
    # kalau mau, bisa ditambah try/except datetime.strptime untuk validasi
    time_str = f"{hour_text.zfill(2)}:{minute_text.zfill(2)}"
    stamp = f"{date_text} {time_str}"
    item_text = f"[{stamp}] {text}"
    listbox_tasks.insert(tk.END, item_text)
    task_var.set("")


btn_add = tk.Button(
    input_frame,
    text="Add",
    font=("Helvetica", 11, "bold"),
    bg="#FA812F",
    fg="black",
    activebackground="#FAB12F",
    activeforeground="black",
    relief="flat",
    command=add_task
)
btn_add.pack(side="right")

# ==== LISTBOX + SCROLLBAR ====
list_frame = tk.Frame(root, bg="#FEF3E2")
list_frame.pack(padx=10, pady=(5, 10), fill="both", expand=True)

scrollbar = tk.Scrollbar(list_frame)
scrollbar.pack(side="right", fill="y")

listbox_tasks = tk.Listbox(
    list_frame,
    font=LIST_FONT,
    bg="#FA812F",
    fg="#000000",
    selectbackground="#FAB12F",
    selectforeground="#000000",
    activestyle="none",
    yscrollcommand=scrollbar.set
)
listbox_tasks.pack(fill="both", expand=True)

scrollbar.config(command=listbox_tasks.yview)

# ==== BUTTONS: DELETE & CLEAR ====
buttons_frame = tk.Frame(root, bg="#FEF3E2")
buttons_frame.pack(pady=10)

def delete_selected():
    selection = listbox_tasks.curselection()
    if selection:
        listbox_tasks.delete(selection[0])

def clear_all():
    listbox_tasks.delete(0, tk.END)

btn_delete = tk.Button(
    buttons_frame,
    text="Delete Selected",
    font=("Helvetica", 11),
    bg="#DD0303",   # red
    fg="black",
    activebackground="#B50202",
    activeforeground="black",
    relief="flat",
    command=delete_selected
)
btn_delete.pack(side="left", padx=5)

btn_clear = tk.Button(
    buttons_frame,
    text="Clear All",
    font=("Helvetica", 11),
    bg="#FAB12F",
    fg="black",
    activebackground="#FA812F",
    activeforeground="black",
    relief="flat",
    command=clear_all
)
btn_clear.pack(side="left", padx=5)

# Optional: Enter key adds task
root.bind("<Return>", lambda event: add_task())

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from datetime import datetime

# ============ SETUP WINDOW ============
root = tk.Tk()
root.title("To-Do List (ttk Version)")
root.geometry("500x500")

# Optional: pilih theme ttk
style = ttk.Style()
style.theme_use("clam")  # bisa coba: 'default', 'clam', 'alt', 'vista' (Windows), dll

# Style dasar untuk beberapa widget
style.configure("Title.TLabel", font=("Helvetica", 20, "bold"))
style.configure("TEntry", font=("Helvetica", 11))
style.configure("TButton", font=("Helvetica", 10))
style.configure("Treeview", font=("Helvetica", 10))
style.configure("Header.TLabel", font=("Helvetica", 10, "bold"))

# ============ TITLE ============
title_label = ttk.Label(root, text="To-Do List", style="Title.TLabel", anchor="center")
title_label.pack(pady=(10, 5))

info_label = ttk.Label(
    root,
    text="Set date & time, type your task, then click Add",
    style="Header.TLabel"
)
info_label.pack(pady=(0, 10))

# ============ DATE & TIME PICKER ============
dt_frame = ttk.Frame(root, padding=(10, 0))
dt_frame.pack(fill="x")

# Date
today_str = datetime.now().strftime("%d-%m-%Y")
date_var = tk.StringVar(value=today_str)

lbl_date = ttk.Label(dt_frame, text="Date (DD-MM-YYYY):")
lbl_date.grid(row=0, column=0, sticky="w")

entry_date = ttk.Entry(dt_frame, textvariable=date_var, width=15)
entry_date.grid(row=0, column=1, sticky="w", padx=(5, 0))

# Time (hour & minute)
lbl_time = ttk.Label(dt_frame, text="Time (HH:MM):")
lbl_time.grid(row=1, column=0, sticky="w", pady=(5, 0))

time_frame = ttk.Frame(dt_frame)
time_frame.grid(row=1, column=1, sticky="w", pady=(5, 0))

hour_var = tk.StringVar(value="08")
minute_var = tk.StringVar(value="00")

spin_hour = ttk.Spinbox(time_frame, from_=0, to=23, width=3, textvariable=hour_var, justify="center")
spin_hour.pack(side="left")

lbl_colon = ttk.Label(time_frame, text=":")
lbl_colon.pack(side="left")

spin_minute = ttk.Spinbox(time_frame, from_=0, to=59, width=3, textvariable=minute_var, justify="center")
spin_minute.pack(side="left", padx=(2, 0))

dt_frame.columnconfigure(1, weight=1)

# ============ ENTRY + ADD BUTTON ============
input_frame = ttk.Frame(root, padding=10)
input_frame.pack(fill="x")

task_var = tk.StringVar()

entry_task = ttk.Entry(input_frame, textvariable=task_var)
entry_task.pack(side="left", fill="x", expand=True, padx=(0, 5))

def add_task(event=None):
    text = task_var.get().strip()
    if not text:
        return

    date_text = date_var.get().strip()
    hour_text = hour_var.get().strip()
    minute_text = minute_var.get().strip()

    # Format jam & menit jadi 2 digit
    time_str = f"{hour_text.zfill(2)}:{minute_text.zfill(2)}"
    stamp = f"{date_text} {time_str}"

    # Masukkan ke Treeview
    tree.insert("", tk.END, values=(stamp, text))
    task_var.set("")

btn_add = ttk.Button(input_frame, text="Add", command=add_task)
btn_add.pack(side="right")

# Enter juga bisa untuk Add
root.bind("<Return>", add_task)

# ============ TREEVIEW (DAFTAR TUGAS) ============
list_frame = ttk.Frame(root, padding=(10, 5))
list_frame.pack(fill="both", expand=True)

columns = ("datetime", "task")
tree = ttk.Treeview(
    list_frame,
    columns=columns,
    show="headings"   # tidak pakai kolom pertama default
)

tree.heading("datetime", text="Date & Time")
tree.heading("task", text="Task")

tree.column("datetime", width=140, anchor="w")
tree.column("task", anchor="w")

# Scrollbar vertikal
scrollbar = ttk.Scrollbar(list_frame, orient="vertical", command=tree.yview)
tree.configure(yscrollcommand=scrollbar.set)

tree.pack(side="left", fill="both", expand=True)
scrollbar.pack(side="right", fill="y")

# ============ BUTTONS: DELETE & CLEAR ============
buttons_frame = ttk.Frame(root, padding=10)
buttons_frame.pack()

def delete_selected():
    selected = tree.selection()
    for item in selected:
        tree.delete(item)

def clear_all():
    for item in tree.get_children():
        tree.delete(item)

btn_delete = ttk.Button(buttons_frame, text="Delete Selected", command=delete_selected)
btn_delete.pack(side="left", padx=5)

btn_clear = ttk.Button(buttons_frame, text="Clear All", command=clear_all)
btn_clear.pack(side="left", padx=5)

root.mainloop()
```
