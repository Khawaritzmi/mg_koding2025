**Pertanyaan:** Bagaimana operator keanggotaan `in` dan `not in` memeriksa daftar?

Cell ini mengecek apakah nilai tertentu terdapat dalam list dan mencetak pesan berbeda berdasarkan hasil pengecekan.

```python
x = 24
y = 20
list = [10, 20, 30, 40, 50]

if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")

if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
```
